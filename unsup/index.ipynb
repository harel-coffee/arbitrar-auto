```python
data, X, clf = utils.isolation_forest("../data/libcrypto_latest_openssl_fopen/features/openssl_fopen", contamination=0.03)
prediction = clf.predict(X)
anomalies = [data[i] for i in range(len(prediction)) if prediction[i] == -1]
anomaly_x = [utils.encode_json(data) for data in anomalies]
anomaly_scores = clf.score_samples(anomaly_x)
zipped = list(zip(anomalies, anomaly_scores))
sorted(zipped, key = lambda d: d[1])
```
